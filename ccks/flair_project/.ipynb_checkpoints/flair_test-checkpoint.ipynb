```python
from flair.data import Corpus
from flair.datasets import UD_ENGLISH
from flair.embeddings import TokenEmbeddings, WordEmbeddings, StackedEmbeddings
```

```python
# init embedding
glove_embedding = WordEmbeddings('zh')# create sentence.
```

```python
from flair.data import Sentence
from flair.models import SequenceTagger
```

```python
sentence = Sentence('今 天 是 中 国 的 生 日 塱 。') #中文ner如果提前分词，可能会有分词错误，导致词性无法标注。所以要按字分,对于embedding里面没有的句子直接按照unk处理了。
```

```python
sentence = Sentence('life is so beatiful alalabitu .')  #句子这一步只会进行简单的分词，根据空格识别有多少个token
```

```python
sentence
```

```python
# embed a sentence using glove.
glove_embedding.embed(sentence)

# now check out the embedded tokens.
for token in sentence:
    print(token)
    print(token.embedding)
```
